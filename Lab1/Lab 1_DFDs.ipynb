```python
## MN Geospatial Commons - Mermaid.js Data Flow Diagram Code

# Work flow
graph TD;
    A[MN Geo Commons] -->|GET request| B[Get Bike Trails Metadata];
    B -->|Convert to JSON| C[JSON Response];
    C -->|Extract URL| D[Extract Bike Trail URL];
    D -->|Download| E[Download Bike Trail Data];
    
    E -->|Check Status| F{Is Status Code 200?};
    F -- Yes --> G[Save as biketrail_data.zip];
    G -->|Unzip if ZIP| H[Unzip File];
    H --> I[Unzipped Content];
    
    F -- No --> J[Failed to Download];

    I -->|Read CSV| K[Read GeoPackage File];
    K -->|Display Data| L[Print DataFrame];

# API conceptual breakdown
graph TD
    A[User Request] -->|GET| B[API Endpoint: package_show]
    B --> C[Parameters]
    C --> C1[id: Dataset ID]
    B --> D[Response]
    D --> E[Success/Failure Status]
    D --> F[Result: Dataset Information]
    F --> F1[Title]
    F --> F2[Description]
    F --> F3[Author]
    F --> F4[Tags]
    F --> F5[Resources]
    
    F5 --> F5a[Resource Name]
    F5 --> F5b[Resource URL]
    F5 --> F5c[Resource Format]
```

```python
## ArcGIS Rest API - Mermaid.js Data Flow Diagram Code

# Work flow
graph TD;
    A[Hennepin Parks URL] -->|Send GET Request| B[parks_request];
    B --> C{Status Code 200?};
    C -- Yes --> D[Save GeoJSON File];
    D --> E[GeoJSON file downloaded successfully];
    C -- No --> F[Failed to download GeoJSON file];

    A2[Twin Cities Counties URL] -->|Send GET Request| B2[response];
    B2 --> C2{Status Code 200?};
    C2 -- Yes --> G[Load GeoJSON Data];
    G --> H[Convert to GeoDataFrame];
    H --> I[Plot GeoDataFrame];
    I --> J[Display Twin Cities Counties Boundaries];
    C2 -- No --> K[Failed to retrieve data];

# API conceptual breakdown
graph TD
    A[User Request] -->|GET| B[API Endpoint: /query]
    B --> C[Parameters]
    
    %% Query Parameters
    C --> C1[where: 1=1]
    C --> C2[text: empty]
    C --> C3[objectIds: empty]
    C --> C4[time: empty]
    C --> C5[timeRelation: esriTimeRelationOverlaps]
    C --> C6[geometry: empty]
    C --> C7[geometryType: esriGeometryEnvelope]
    C --> C8[inSR: empty]
    C --> C9[spatialRel: esriSpatialRelIntersects]
    C --> C10[outFields: all fields]
    C --> C11[returnGeometry: true]
    C --> C12[returnTrueCurves: false]
    C --> C13[f: geojson]

    %% Response
    B --> D[Response: GeoJSON]
    D --> E[Features]
    E --> E1[Geometry: Coordinates]
    E --> E2[Properties: Feature Attributes]
```

```python
## Spatial Join - Mermaid.js Data Flow Diagram Code

graph TD;
    A[Load GeoPackage Layers] -->|List Layers| B[gpkg_layers];
    B --> C[Load Bike Trails GeoDataFrame];
    C --> D[Load Parks GeoDataFrame];
    
    D --> E[Check CRS of GeoDataFrames];
    E --> F{Do CRS Match?};
    F -- No --> G[Reproject Parks CRS to Match Bike Trails];
    G --> H{Do Counties CRS Match?};
    H -- No --> I[Reproject Counties CRS to Match Bike Trails];
    
    E --> J[Plot GeoDataFrames];
    
    J --> K[Spatial Join Bike Trails with Counties];
    K --> L[Bike Trails within Twin Cities Boundaries];
    
    J --> M[Spatial Join Parks with Counties];
    M --> N[Parks within Twin Cities Boundaries];
    
    L --> O[Plot Results];
    N --> O;
    
    O --> P[Display Title and Legend];
    P --> Q[Print Heads of Merged GeoDataFrames];
```
