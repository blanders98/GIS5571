```python
## ArcGIS Rest API - Mermaid.js Data Flow Diagram Code

# Work flow
graph TD;
    A[Hennepin Parks URL] -->|Send GET Request| B[parks_request];
    B --> C{Status Code 200?};
    C -- Yes --> D[Save GeoJSON File];
    D --> E[GeoJSON file downloaded successfully];
    C -- No --> F[Failed to download GeoJSON file];

    A2[Twin Cities Counties URL] -->|Send GET Request| B2[response];
    B2 --> C2{Status Code 200?};
    C2 -- Yes --> G[Load GeoJSON Data];
    G --> H[Convert to GeoDataFrame];
    H --> I[Plot GeoDataFrame];
    I --> J[Display Twin Cities Counties Boundaries];
    C2 -- No --> K[Failed to retrieve data];

# API conceptual breakdown
graph TD
    A[User] -->|Sends Query Request| B[API Endpoint]
    B -->|Query Parameters: where=1=1, returnGeometry=true, outFields=*, etc.| C[Query Layer 3 - PLACES]

    C --> D[API Processes Query]
    
    D -->|If Valid Request| E[Returns GeoJSON Data]
    D -->|If Invalid| F[Returns Error Message]

    E --> G[GeoJSON Data: Feature Collection with Geometries and Attributes]
    G --> H[User Downloads GeoJSON]

    F -->|Error Handling| K[User Receives Error Message]
```
